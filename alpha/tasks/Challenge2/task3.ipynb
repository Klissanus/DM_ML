Блокбастер, готовящийся к выходу в 2017 году, собрал 198 положительных и 38 отрицательных отзывов, из них 27 сильно положительные и 14 сильно отрицательные. Предскажите выручку от этого фильма с минимальной относительной ошибкой.  
Воспользуйтесь данными http://www.stat.ufl.edu/~winner/data/box_office.xls. Кратко опишите свой подход. 

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_excel('box_office.xls')
df = df.drop(['Rank', 'Movie', 'Release'], axis = 1)
df
```

```python
x_p = np.array([198, 38, 27, 14, 2017]).reshape(1, -1)
```

```python
x = df.drop(['Revenues'], 'columns')
y = df.Revenues
```

```python
lr = LinearRegression()
lr.fit(x, y)
lr.predict(x_p)
```

```python
rfr = RandomForestRegressor()
rfr.fit(x, y)
rfr.predict(x_p)
```
